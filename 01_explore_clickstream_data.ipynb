## Build A Product Co-Occurence Table

```python
#Get the set of products per session

session_products = events.groupby("session_id")["product_id"].apply(lambda x: set(x.dropna()))
session_products.head()
```

```python
## shows which products were touched in each session. "users who touched A also touched B" foundation
```

```python
#build the co-ccurence pairs 

from itertools import combinations
from collections import Counter 

pair_counts = Counter ()

for products in session_products:
    for combo in combinations(products, 2): 
        pair_counts[tuple(sorted(combo))] +=1

len(pair_counts) 
```

```python
#Tuen it into a DataFrame

co_occurrence_df = pd.DataFrame(
    [(p1, p2, count) for (p1, p2), count in pair_counts.items()],
    columns=["product_1", "product_2", "co_count"]
).sort_values("co_count", ascending=False)

co_occurrence_df.head(10)
```
